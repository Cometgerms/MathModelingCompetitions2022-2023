```python
'''
This program is used to fit the data of the total number of people and the total number of eggs present int the bee hive
Used in: HiMCM 2022 Problem 1
'''

'''
Fitting of the total number of people
'''

from scipy.interpolate import CubicSpline
import numpy as np

import matplotlib.pyplot as plt
y_year = []
x = [1, 11, 21, 32, 42, 52, 60, 70, 80, 91, 101, 111, 121, 131, 141, 152, 162, 172, 182, 192,202,213,223,233,244,254,264,274,284,294,305,315,325,335,345,355, 366]
p = [11800, 16150, 21620, 27430, 34545, 41015, 48605, 53935, 61335, 65980, 65325, 66285, 63760, 61015, 60370, 57915, 54015, 46765, 42895, 39230, 31815, 32660, 35135, 36990, 37855, 37690, 35005, 32875, 31175, 27665, 23070, 17270, 14560, 11895, 10005, 7345, 11800]
f = CubicSpline(x, p, bc_type='natural')
x_year = np.linspace(1, 365, 365)
for i in range(0,365):
    y_year.append(f(i+1))

plt.figure(figsize = (10,8))
plt.plot(x_year, y_year, )
# plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation', fontsize = 20)
plt.xlabel('x')
plt.ylabel('y')


plt.show()

print('[')
for i in range(0,365):
    print(y_year[i] , ",")
print(']')
```

```python
'''
Fitting of the total number of eggs present
'''


x = [1, 11, 21, 32, 42, 52, 60, 70, 80, 91, 101, 111, 121, 131, 141, 152, 162, 172, 182, 192,202,213,223,233,244,254,264,274,284,294,305,315,325,335,345,355, 366]
egg = [1315, 1915, 1735, 2815, 2835, 3225, 3695, 3045, 3125, 2780, 2975, 2929, 2995, 2950, 2925, 1590, 1595, 650, 1870, 1185, 2270, 1545, 1970, 1505, 1700, 1215, 1425, 1055, 1650, 0, 0, 470, 380, 725, 365, 1015, 1315]
f = CubicSpline(x, egg, bc_type='natural')
x_year_egg = np.linspace(1, 365, 365)
y_egg_total = f(x_year)

for i in range (0,365):
    if(y_egg_total[i]<0):
        y_egg_total[i]=0
        
plt.figure(figsize = (10,8))
plt.plot(x_year_egg, y_egg_total, 'b')
# plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
print(y_egg_total)
```

```python
'''
Fitting of daily egg laying data
'''
sum_lay_egg=0
lay_egg = []
for i in range (0, 365):
    lay_egg.append(0)
lay_egg[305]=66.12717

for i in range (306, 365):
    lay_egg[i]=y_egg_total[i]-0.96*lay_egg[i-1]-0.96*lay_egg[i-2]
    if(lay_egg[i]<0):
        lay_egg[i]=0
        
lay_egg[0]=y_egg_total[0]-0.96*lay_egg[364]-0.96*lay_egg[363]
lay_egg[1]=y_egg_total[1]-0.96*lay_egg[0]-0.96*lay_egg[364]
lay_egg[2]=y_egg_total[2]-0.96*lay_egg[1]-0.96*lay_egg[0]

for i in range (3, 305):
    lay_egg[i]=y_egg_total[i]-0.96*lay_egg[i-1]-0.96*lay_egg[i-2]
    if(lay_egg[i]<0):
        lay_egg[i]=0
    sum_lay_egg+=lay_egg[i]
    
plt.figure(figsize = (10,8))
plt.plot(x_year_egg, lay_egg, '#006471')
# plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation', fontsize = 25)
plt.xlabel('W', fontsize = 15)
plt.ylabel('Number of Egg laid', fontsize = 15)
plt.grid()
plt.savefig('Egg Laying diagram1', dpi=300)
plt.show()
print(lay_egg)
print(sum_lay_egg)
```
