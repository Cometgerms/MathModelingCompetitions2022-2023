```python
'''
This program is for modelling temperature impact on egg laying
Used in: HiMCM 2022 Problem 1
'''

#x-axis label: temperature (degrees Celsius)
#y-axis label: Number of eggs laid
#Highest point of the mark (ps/direct use of the mark): 113.5


import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
import pandas as pd
import xlrd

'''
modelling temperature impact on egg laying
'''
plt.figure(figsize=(10,6), dpi=300)
temp_sample = [73, 76, 79, 82, 85, 88, 91, 94]
egg_lay = [0, 2, 19, 34, 67, 110, 96, 22]
f_tempidx_egg = CubicSpline(temp_sample, egg_lay, bc_type='natural')

x = np.linspace(0,35,100)
y=[]
for i in range(0,len(x)):
    y.append(f_tempidx_egg(x[i]*1.8+32))
    
    if(y[i]<0):
        y[i]=0

plt.plot(x,y,color='#EBA912')
# plt.title('Temp Egg Index Diagram', fontsize=20)

plt.xlabel('Temperature', fontsize = 15)
plt.ylabel('Number of Egg laid', fontsize = 15)
plt.grid()
plt.savefig('Temp_Egg_Graph.jpg', dpi=300)
plt.show()
```
